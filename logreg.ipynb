```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
```

```
#X , y = make_classification(n_samples=100,n_features=1, n_redundant=0, n_informative=1,
                            # n_clusters_per_class=1)
#plt.scatter(X,y)
```

```
X, Y = make_classification(n_features=2, n_redundant=0, n_informative=2,
                             n_clusters_per_class=1)
plt.scatter(X[:, 0], X[:, 1], marker='o',
            c=Y,
            s=25)
plt.show()
```

```
x0 = np.ones([100,1])
X = np.hstack((x0,X))
```

```
alpha = lr = 0.001
m = len(X)
n = len(list(zip(*X)))
epochs = 2000
theta = np.zeros([n,1])     #This line initializes the value for thetas/parameters to zero
y = np.reshape(Y,[100,1])   #since output y is not in a perfect matrix form we will reshape y to a matrix of [100 x 1]
```

```
def costfunc(theta,X,y):
  temp = np.dot(X,theta)
  temp = 1/(1 + np.exp(-temp))
  return np.sum((y*np.log(temp))+((1 - y)*np.log(1-temp)))*(-1/m)
```

```
costfunc(theta,X,y)
```

```
zeros = np.zeros([100,1])
plt.scatter(X[:, 1], X[:, 2], marker='o',
            c=Y,
            s=25)

print('for theta initialized as zero we get the decion boundary as below')
plt.plot(X[:, 1],zeros)
plt.show()
```

```
J = np.empty([epochs,1])
sqrer = iter(J)
for i in range(epochs):
  temp = np.dot(X,theta)
  temp = (1/(1 + np.exp(-temp))) - y
  temp = np.dot(X.T,temp)
  theta = theta - (lr * temp)
  J[i] = costfunc(theta,X,y)
print ('theta values are:',theta)
print ('error after final iteration is :',J[epochs-1])
```

now let us put a decision boundary along the data based on the theta values we got

```
#
X2 = (-(theta[0]/theta[2]))-(theta[1]/theta[2])*X[:,1]
```

```
plt.scatter(X[:, 1], X[:, 2], marker='o',
            c=Y,
            s=25)
plt.plot(X[:, 1],X2)
plt.show()
```
